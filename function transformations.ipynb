```
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
import plotly.express as px
import plotly.graph_objects as go
```

```
x = np.linspace(-1,2)
```

```
def f(x):
  return 6.75 * ((x ** 3) - (2*(x**2)) + (x))
```

```
a = [f(i) for i in x]
b = [f(i+2) for i in x]
c = [f(i)+2 for i in x]
d = [f(i)*2 for i in x]
e = [f(2*i) for i in x]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=a,
    mode='lines+markers',
    name='f(x)'
))

fig.add_trace(go.Scatter(
    x=x,
    y=b,
    mode='lines+markers',
    name='f(x+2)'
))

fig.add_trace(go.Scatter(
    x=x,
    y=c,
    mode='lines+markers',
    name='f(x)+2'
))

fig.add_trace(go.Scatter(
    x=x,
    y=d,
    mode='lines+markers',
    name='2f(x)'
))

fig.add_trace(go.Scatter(
    x=x,
    y=e,
    mode='lines+markers',
    name='f(2x)'
))

fig.show()
```
